### How to know the crop that has the most consistent yield over time

```python
consistency_metrics = farm_cleaned.groupby('Type of crop')['Average yield (kilograms per hectare)'].agg([
    ('mean_yield', 'mean'),
    ('std_dev', 'std'),
    ('cv', lambda x: x.std() / x.mean() * 100),  # Coefficient of Variation
    ('min_yield', 'min'),
    ('max_yield', 'max'),
    ('range', lambda x: x.max() - x.min()),
    ('data_points', 'count')
]).round(2)
```

```python
consistency_metrics
```

```python
most_consistent = consistency_metrics.sort_values('cv')
```

```python
most_consistent
```

```python
print("Crop Yield Consistency Analysis (1908-1926)")
print("=" * 80)
print("Most Consistent Crops (Lowest Coefficient of Variation):")
print("=" * 80)
for i, (crop, metrics) in enumerate(most_consistent.head(10).iterrows()):
    print(f"{i+1:2d}. {crop:<30} | CV: {metrics['cv']:6.1f}% | Mean: {metrics['mean_yield']:8,.0f} kg/ha | Std: {metrics['std_dev']:8,.0f}")
```

```python
print("\n" + "=" * 80)
print(f"üèÜ MOST CONSISTENT CROP: {most_consistent.index[0]}")
print(f"   Coefficient of Variation: {most_consistent.iloc[0]['cv']:.1f}%")
print(f"   Standard Deviation: {most_consistent.iloc[0]['std_dev']:,.0f} kg/ha")
print(f"   Mean Yield: {most_consistent.iloc[0]['mean_yield']:,.0f} kg/ha")
print(f"   Yield Range: {most_consistent.iloc[0]['min_yield']:,.0f} - {most_consistent.iloc[0]['max_yield']:,.0f} kg/ha")
```

```python
print("\n" + "=" * 80)
print("Least Consistent Crops (Highest Coefficient of Variation):")
print("=" * 80)
for i, (crop, metrics) in enumerate(most_consistent.tail(5).iterrows()):
    print(f"{crop:<30} | CV: {metrics['cv']:6.1f}%")
```

### Additional analysis: Year-over-year changes

```python
print("\n" + "=" * 80)
print("Year-over-Year Yield Stability Analysis:")
print("=" * 80)
yoy_stability = {}
for crop in farm_cleaned['Type of crop'].unique():
    crop_data = farm_cleaned[farm_cleaned['Type of crop'] == crop].sort_values('REF_DATE')
    if len(crop_data) > 1:
        crop_data = crop_data.dropna(subset=['Average yield (kilograms per hectare)'])
        crop_data['yoy_change'] = crop_data['Average yield (kilograms per hectare)'].pct_change() * 100
        avg_absolute_change = crop_data['yoy_change'].abs().mean()
        yoy_stability[crop] = avg_absolute_change
        yoy_stability_sorted = pd.Series(yoy_stability).sort_values()

print("\nMost Stable Crops (Lowest Average Year-over-Year Change):")
for crop, change in yoy_stability_sorted.head().items():
    print(f"  {crop:<30}: {change:.1f}% average YoY change")
```

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
crop_features = farm_cleaned.groupby('Type of crop').agg({
    'Average farm price (dollars per tonne)': 'mean',
    'Average yield (kilograms per hectare)': 'mean',
    'Production (metric tonnes)': 'mean',
    'Seeded area (hectares)': 'mean'
}).fillna(0)
```

```python
crop_features
```

```python
scaler = StandardScaler()
```

```python
scaler
```

```python
scaled_features = scaler.fit_transform(crop_features)
```

```python
scaled_features
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
```

```python
kmeans
```

```python
crop_features['Cluster'] = kmeans.fit_predict(scaled_features)
```
